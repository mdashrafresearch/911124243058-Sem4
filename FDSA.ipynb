```python
import nltk
from nltk.corpus import brown

# Make sure the Brown corpus is downloaded
nltk.download('brown')

# Build a Conditional Frequency Distribution:
# condition = genre, event = word
cfd = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)

# List all conditions (genres)
cfd.conditions()

# Tabulate frequencies of selected words in selected genres
cfd.tabulate(
    conditions=['government', 'humor', 'reviews'],
    samples=['leadership', 'worship', 'hardship']
)

# Plot cumulative frequencies
cfd.plot(
    conditions=['government', 'humor', 'reviews'],
    samples=['leadership', 'worship', 'hardship'],
    cumulative=True
)

# Access the frequency distribution for the 'news' genre
news_fd = cfd['news']

# Most common 3 words in 'news'
news_fd.most_common(3)

# Frequency of the word 'the' in 'news'
news_fd['the']
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(10)
x_random = np.arange(0, 60)
y_random = x_random + np.random.randint(5, 25, 60)

# -------------------------------
# Simple Linear Regression (Manual Calculation)
# -------------------------------

# Calculate slope (m) and intercept (c)
x_mean = np.mean(x_random)
y_mean = np.mean(y_random)

numerator = np.sum((x_random - x_mean) * (y_random - y_mean))
denominator = np.sum((x_random - x_mean) ** 2)

m = numerator / denominator
c = y_mean - m * x_mean

# Predicted line
y_pred = m * x_random + c

# -------------------------------
# Plot Scatter with SLR Line
# -------------------------------

plt.figure(figsize=(8,5))
plt.scatter(x_random, y_random, label="Data Points")
plt.plot(x_random, y_pred, color='red', linewidth=2, label="SLR Line")
plt.title("Random Scatter Plot with Simple Linear Regression Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Z-Test Case Study

# Step 1 & 2: Import required libraries
import math
import numpy as np
from numpy.random import randn
from statsmodels.stats.weightstats import ztest

# Step 3: Define parameters
mean_iq = 110
sample_size = 50
population_sd = 15
alpha = 0.05
null_mean = 100

# Standard Error
standard_error = population_sd / math.sqrt(sample_size)

# Generate sample data
data = standard_error * randn(sample_size) + mean_iq

# Print sample statistics
print("Sample Mean = %.2f" % np.mean(data))
print("Sample Standard Deviation = %.2f" % np.std(data))

# Step 4: Perform Z-test
z_score, p_value = ztest(data, value=null_mean, alternative='larger')

print("Z-Score =", z_score)
print("P-Value =", p_value)

# Step 5: Decision Making
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
# Importing required libraries
import numpy as np
from scipy import stats

# Sample Size
N = 10

# Generate random samples
# Gaussian distributed data with mean = 2 and variance = 1
x = np.random.randn(N) + 2

# Gaussian distributed data with mean = 0 and variance = 1
y = np.random.randn(N)

# -------------------------------
# Manual T-Test Calculation
# -------------------------------

# Sample variances
var_x = x.var(ddof=1)
var_y = y.var(ddof=1)

# Pooled Standard Deviation
SD = np.sqrt((var_x + var_y) / 2)

print("Pooled Standard Deviation =", SD)

# T-statistic calculation
tval = (x.mean() - y.mean()) / (SD * np.sqrt(2 / N))

# Degrees of freedom
dof = 2 * N - 2

# Two-tailed p-value
pval = 2 * (1 - stats.t.cdf(abs(tval), df=dof))

print("Manual Calculation")
print("t-value =", tval)
print("p-value =", pval)

# -------------------------------
# Cross-check using SciPy function
# -------------------------------

tval2, pval2 = stats.ttest_ind(x, y)

print("\nUsing scipy.stats.ttest_ind()")
print("t-value =", tval2)
print("p-value =", pval2)
```

```python
# Import required libraries
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

# Sample Data (Example)
data = {
    'A': [25, 30, 28, 32, 27],
    'B': [35, 40, 36, 38, 39],
    'C': [45, 42, 48, 50, 46],
    'D': [20, 22, 19, 24, 21]
}

df = pd.DataFrame(data)

# -------------------------------
# 1. Using scipy.stats.f_oneway()
# -------------------------------

fvalue, pvalue = stats.f_oneway(df['A'], df['B'], df['C'], df['D'])

print("F-value:", fvalue)
print("P-value:", pvalue)


# -------------------------------
# 2. ANOVA Table using statsmodels (R-like output)
# -------------------------------

# Convert wide format to long format
df_melt = df.melt(var_name='treatments', value_name='value')

# Ordinary Least Squares (OLS) Model
model = ols('value ~ C(treatments)', data=df_melt).fit()

anova_table = sm.stats.anova_lm(model, typ=2)

print("\nANOVA Table:")
print(anova_table)
```

```python
# -----------------------------------
# IMPORT REQUIRED LIBRARIES
# -----------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

warnings.filterwarnings('ignore')

# -----------------------------------
# STEP 1: GENERATE HOUSING DATASET
# -----------------------------------
np.random.seed(42)

n = 545  # number of samples

area = np.random.randint(500, 5000, n)
bedrooms = np.random.randint(1, 6, n)
bathrooms = np.random.randint(1, 4, n)
stories = np.random.randint(1, 4, n)
parking = np.random.randint(0, 3, n)

price = (area * 2500 +
         bedrooms * 50000 +
         bathrooms * 40000 +
         stories * 30000 +
         parking * 20000 +
         np.random.randint(-100000, 100000, n))

housing = pd.DataFrame({
    'area': area,
    'bedrooms': bedrooms,
    'bathrooms': bathrooms,
    'stories': stories,
    'parking': parking,
    'price': price
})

# Save dataset
housing.to_csv("Housing.csv", index=False)

print("Dataset Created Successfully")
print(housing.head())

# -----------------------------------
# STEP 2: LOAD DATASET
# -----------------------------------
housing = pd.read_csv("Housing.csv")

print("\nNumber of Observations:", len(housing))

# -----------------------------------
# STEP 3: SELECT AREA & PRICE
# -----------------------------------
df = housing[['area', 'price']].copy()

# -----------------------------------
# STEP 4: SCALE VARIABLES
# -----------------------------------
scaler = MinMaxScaler()
df[['area', 'price']] = scaler.fit_transform(df[['area', 'price']])

# -----------------------------------
# STEP 5: VISUALIZATION
# -----------------------------------
sns.regplot(x="area", y="price", data=df, fit_reg=False)
plt.title("Area vs Price (Scaled Data)")
plt.show()

# -----------------------------------
# STEP 6: TRAIN-TEST SPLIT
# -----------------------------------
df_train, df_test = train_test_split(
    df,
    train_size=0.7,
    test_size=0.3,
    random_state=10
)

print("Training samples:", len(df_train))
print("Testing samples:", len(df_test))

# -----------------------------------
# STEP 7: PREPARE X AND y
# -----------------------------------
X_train = df_train['area'].values.reshape(-1, 1)
y_train = df_train['price']

X_test = df_test['area'].values.reshape(-1, 1)
y_test = df_test['price']

# -----------------------------------
# STEP 8: POLYNOMIAL REGRESSION MODEL
# -----------------------------------
degree = 2
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

model.fit(X_train, y_train)

# -----------------------------------
# STEP 9: MODEL EVALUATION
# -----------------------------------
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print("\nTraining R Score:", train_score)
print("Testing R Score:", test_score)

# -----------------------------------
# STEP 10: CROSS VALIDATION
# -----------------------------------
kfold = KFold(n_splits=5, shuffle=True, random_state=10)
cv_scores = cross_val_score(model, X_train, y_train, cv=kfold, scoring='r2')

print("\nCross Validation R Scores:", cv_scores)
print("Mean CV Score:", np.mean(cv_scores))

# -----------------------------------
# STEP 11: PLOT REGRESSION CURVE
# -----------------------------------
x_range = np.linspace(df['area'].min(), df['area'].max(), 100).reshape(-1,1)
y_pred = model.predict(x_range)

plt.scatter(X_train, y_train, label="Training Data")
plt.plot(x_range, y_pred, color='red', linewidth=2, label="Polynomial Fit")
plt.title("Polynomial Regression Curve")
plt.xlabel("Area")
plt.ylabel("Price")
plt.legend()
plt.show()
```
