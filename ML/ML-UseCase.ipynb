```python
# Install pgmpy
!pip install pgmpy -q

# Import libraries
import pandas as pd
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination

# -----------------------------
# Step 1: Create Dataset
# -----------------------------
data = {
    'StudyHours': ['High', 'High', 'Low', 'Low', 'High', 'Low'],
    'Attendance': ['High', 'Low', 'High', 'Low', 'High', 'Low'],
    'Pass': ['Yes', 'Yes', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)

print("Dataset:")
display(df)

# -----------------------------
# Step 2: Define Bayesian Network
# -----------------------------
model = DiscreteBayesianNetwork([
    ('StudyHours', 'Pass'),
    ('Attendance', 'Pass')
])

# -----------------------------
# Step 3: Train Model
# -----------------------------
model.fit(df, estimator=MaximumLikelihoodEstimator)

# -----------------------------
# Step 4: Inference
# -----------------------------
inference = VariableElimination(model)

# Example Query
evidence = {
    'StudyHours': 'Low',
    'Attendance': 'Low'
}

result = inference.query(variables=['Pass'], evidence=evidence)

print("\nProbability of Passing (Low Study, Low Attendance):")
print(result)

# MAP prediction
map_result = inference.map_query(variables=['Pass'], evidence=evidence)

print("\nMost Likely Outcome:")
print(map_result)
```
